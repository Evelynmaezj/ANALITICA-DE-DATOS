Eliminar filas por medios de skiprows

```python
#Eliminar primera fila o filas innecesarias
data1= pd.read_csv('inmuebles2.csv', skiprows=1)
data1
```

Eliminar columnas por DROP


```python
#axis = 1 columnas
#axis = 0 filas
data2 = data1.drop(['0','Referencia'], axis=1)
data2
```

```python
data2.info()
```

```python
#Identificar valores nulos por columna
valores_nulos = data2.isnull().sum()
valores_nulos
```

Valores nulos por dataframe

```python
#Identificar valores nulos por data frame
valores_nulos = data2.isnull().sum().sum()
valores_nulos
```

##**Métodos de sustitución de datos nulos**

```python
#Realizamos una copia del dataframe
data3 = data2.copy()
```

###1.Sustitución con promedio

```python
#Primer método de sustitución de valores nulos
#Sustituir valores nulos con promedio o media
data3["Superficie"]=data3['Superficie'].fillna(round(data2['Superficie'].mean(),1)) #,1 es el # de decimales
```

```python
#Comprobar valores nulos limpios
valores_nulos=data3.isnull().sum()
valores_nulos
```

###2.Sustitución con mediana


```python
#Segundo método de sustitución de valores nulos
#Sustituir valores nulos con mediana
data3['Precio Venta'] = data3['Precio Venta'].fillna(round(data3['Precio Venta'].median(), 1))
```

```python
#Comprobar valores nulos limpios
valores_nulos=data3.isnull().sum()
valores_nulos
```

###3.Sustitución con un valor en concreto


```python
#Tercer método de sustitución de valores nulos
#Sustituir valores nulos por un valor en especifico
data3["Fecha Alta"]= data3["Fecha Alta"].fillna(10)
```

```python
#Comprobar valores nulos limpios
valores_nulos=data3.isnull().sum()
valores_nulos
```

###4.Sustitución con un string en concreto

```python
#Cuarto método de sustitución de valores nulos
#Sustituir valores nulos por un string en especifico
data3["Fecha Venta"]= data3["Fecha Venta"].fillna("2023-06-19 00:00:00")
```

```python
#Comprobar valores nulos limpios
valores_nulos=data3.isnull().sum()
valores_nulos
```

###5.Sustitución Forward fill

Si el primero es nulo no lo va a rellenar, y si hay 3 nulos seguidos solo va a rellenar el primero

```python
#Quinto método de sustitución de valores nulos
#Sustituir valores nulos por valores ni nulos hacia adelante "forward fill" ("ffill")
#Filtro por columnas
data3["Tipo"]= data3["Tipo"].fillna(method="ffill")
```

```python
#Comprobar valores nulos limpios
valores_nulos=data3.isnull().sum()
valores_nulos
```

###6.Sustitución Backward fill

```python
#Sexto método de sustitución de valores nulos
#Sustituir valores nulos por valores ni nulos hacia atrás "backwaard fill" ("bfill")
#Filtro por columnas
data3["Operación"]= data3["Operación"].fillna(method="bfill")
```

```python
#Comprobar valores nulos limpios
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Sustitución de los demás valores nulos
data3["Provincia"]= data3["Provincia"].fillna(method="bfill")
data3["Vendedor"]= data3["Vendedor"].fillna(method="ffill")
```

```python
#Comprobar valores nulos limpios
valores_nulos=data3.isnull().sum()
valores_nulos
```
