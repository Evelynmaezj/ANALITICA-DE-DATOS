##Filtros básicos para variables cualitativas


####Filtro por objeto

```python
filtro1=df[df["NOMBRE_CLIENTE"]=="PÚBLICO EN GENERAL"]
filtro1

#filtro1_2=df[df["CVE_DOC"]=="F37036"]
#filtro1_2

#filtro1_3=df[df["CANT"]==7]
#filtro1_3
```

####Filtro por filas

```python
#[FILAS , COLUMNAS]
#filtro2 = df.iloc[8:15,: ]  #Filas consecutivas
filtro2 = df.iloc[[8, 24, 50] , :]  #Filas no consecutivas
filtro2
```

####Filtro por columnas

```python
filtro3 = df.iloc[: , 1:3]  #Columnas consecutivas
filtro3 = df.iloc[:,[8, 11]]   #Columnas no consecutivas
filtro3
```

####Filtro de filas vs columnas

```python
#Paso 1: Reindexar
df1 = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=0)
df1
```

```python
#Paso 2: Aplicamos filtro filas vs columnas
filtro4 = df1.loc[["F37034", "F39838"], ["COSTO_TOTAL_CALCULADO"]]
filtro4
```

####Filtro por cabecera

```python
filtro5 = df.tail(2) #()indicar cuantas filas mostrar
filtro5
```

##Filtros básicos para variables cuantitativas

####Filtro de comparación

```python
filtro6 = df[df["COSTO_TOTAL_CALCULADO"]>= 10000]
filtro6
```

##Filtros básicos con conectores lógicos

####Filtro Y

```python
filtro7 = df[(df["COSTO_TOTAL_CALCULADO"] < 10000) & (df["FECHA_DOC"]=="2022-03-14")]
filtro7
```

####Filtro O

```python
filtro8 = df[(df["COSTO_TOTAL_CALCULADO"] < 10000) | (df["FECHA_DOC"]=="2022-03-14")]
filtro8
```

####Filtro Not

```python
#filtro9 = df[~(df["FECHA_DOC"]=="2022-03-14")]
filtro9 = df[~(df["COSTO_TOTAL_CALCULADO"] < 1000) & (df["FECHA_DOC"]=="2022-12-23")]
filtro9
```
