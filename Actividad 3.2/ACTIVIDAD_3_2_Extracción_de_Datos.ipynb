Limpieza de datos

```python
#Reemplazo de datos tipo strinf a datos tipo númerico
df.CVE_CLPV = df.CVE_CLPV.replace({'O':'0'}, regex=False) #regex significa alg regresivo hace un recorrido regresivo para evitar que vuelva al inicio y siga en donde estabamos
df.CVE_CLPV = df.CVE_CLPV.replace({'6781|':'6781'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'3612C':'3612'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'      ':'0'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'2385A':'2385'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'2510A':'2510'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'8440Q':'8440'}, regex=False)
df.CVE_CLPV = df.CVE_CLPV.replace({'M':'0'}, regex=False)
```

```python
#Conversión de tipo Obj a Int
df["CVE_CLPV"] = df["CVE_CLPV"] .astype(int)
```

```python
#Ajustar el máximo de filas o columnas
#pd.options.display.max_rows = None
#pd.options.display.max_columns = None
pd.options.display.max_rows = 10
pd.options.display.max_columns = None
```

####Filtro de facturas con clave entre 1000 y 2000



```python
filtro1 = df[(df["CVE_CLPV"] > 1000) & (df["CVE_CLPV"] < 2000)]
filtro1
```

####Filtro de todas las claves excepto 5 y 4

```python
filtro2 = df[~(df["CVE_VEND"] == 5) & ~(df["CVE_VEND"] == 4)]
filtro2
```
