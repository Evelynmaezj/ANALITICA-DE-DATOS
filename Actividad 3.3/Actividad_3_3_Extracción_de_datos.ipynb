##a) Eliminar la primera fila
Ya que esta fila no contiene información relevante, vuelve a leer el data frame y se agrega con una coma skiprows para inficar el número de fila iniciando por 1

```python
df=pd.read_excel('Inmuebles.xlsx', skiprows=1)
df
```

##a) Eliminar la columnas
Por medio de drop, usando:

*   axis=0 para filas
*   axis=1 para  columnas

Se eliminaran la columna "0" y "Referencia"  

```python
#Eliminar columnas innecesarias
#axis = 1 para columnas
#axis = 0 para filas
#df=pd.read_excel('Inmuebles.xlsx',skiprows=1, index_col=1)
df = df.drop(['Referencia'], axis=1)
df
```

##b) Filtro de registros "Fecha_Alta"
Fecha Alta : igual a 2007-04-29 00:00:00 y a 2004-01-02 00:00:00

```python
filtro1 = df[(df['Fecha_Alta'] == '2007-04-29') | (df['Fecha_Alta'] =='2004-01-02')]
filtro1
```

##c) Filtro de registros en tipo
Todos los registros de “ Tipo: parking, local, industrial y piso”

```python
filtro2 = df[(df['Tipo'] == 'Parking') |
             (df['Tipo'] == 'Local') |
             (df['Tipo'] == 'Suelo') |
             (df['Tipo'] =='Piso')]
filtro2
```

##d) Filtro de Provincia **Y** Superficie
Todos los registros con “Provincia” igual a “Girona” y con “Superficie” mayor a “100”

```python
filtro3 = df[(df['Provincia'] == 'Girona') &
             (df['Superficie'] > 100)]
filtro3
```

##e) Filtro Precio Venta **O** Vendedor
Todos los registros con “Precio Venta: mayor a 1000,000” o “vendedor” igual a “Carmen o Joaquin”

```python
filtro4 = df[(df['Precio Venta'] > 1000000) |
             (df['Vendedor'] =='Carmen') |
             (df['Vendedor'] =='Joaquin')]
filtro4
```

##f) Filtro negación de Provincia
Todos los registros que no sean de “Provincia : Barcelona“

```python
filtro5 = df[~(df['Provincia'] == 'Barcelona')]
filtro5
```

##g) Filtro negación de Operación y Tipo
Todos los registros que no sean de “Operación: alquiler “ ni “Tipo: oficina“

```python
filtro6 = df[~(df['Operación'] == 'Alquiler') &
               ~(df['Tipo'] == 'Oficina')]
filtro6
```

##Guardado de filtros y archivo en CSV

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("Entregable1.csv")
filtro2.to_csv("Entregable2.csv")
filtro3.to_csv("Entregable3.csv")
filtro4.to_csv("Entregable4.csv")
filtro5.to_csv("Entregable5.csv")
filtro6.to_csv("Entregable6.csv")
```
